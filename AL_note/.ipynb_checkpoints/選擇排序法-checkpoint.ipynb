```python
import random
data=[]
for x in range(1,101):
    data.append(x)
random.shuffle(data)
```

```python
for x in range(len(data)): #讓一個指定的暫時變數從0號開始跑，直到餵進去的資料長度後停止。
    min_num = x
    for y in range(x+1, len(data)):   #不斷把x跟x裡面的其他數(y)作對比，跑到x的長度結束(全部比較一遍)，比較大就覆蓋，讓數字成長。
        if data[min_num]> data[y]:    #range(從x+1到末數)
            min_num = y               #如果比較小就不動作，會繼續比
    data[x], data[min_num] = data[min_num], data[x] #執行交換動作
#[12,24,14,28,52]->[12, 14, 24, 28, 52]跑一次的結果-->12跟全部比都最小，所以min_num始終是0號，所以交換完還是一樣。
#跑到這邊後會回去第一層迴圈，從1號繼續跑下去。然後24比14大，所以min_num就會被覆蓋變成用14往下比，比完就交換位子。
#第二次結果是24跟14會交換位置。
for z in range(len(data)):
    print(data[z]) #print出最終結果，最上面迴圈跑完才會跑這個，因為級別一樣大會看前後順序
```
