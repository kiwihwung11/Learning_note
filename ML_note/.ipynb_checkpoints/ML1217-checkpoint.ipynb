```python
#創造半月型數據
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
X,y=make_moons(n_samples=200,noise=0.05,random_state=0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
#用DBSCAN進行訓練
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.2,min_samples=5,metric="euclidean")
y_db=db.fit_predict(X)
```

```python
#畫圖
plt.scatter(X[y_db == 0,0],
            X[y_db == 0,1],
            s=50,c='lightgreen',
            marker='s',edgecolor='black',
            label='cluster 1')
plt.scatter(X[y_db == 1,0],
            X[y_db == 1,1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='cluster 2')
plt.scatter(X[y_db == 2,0],
            X[y_db == 2,1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='cluster 3')
#plt.scatter(labels.cluster_centers_[:, 0],
            #labels.cluster_centers_[:, 1],
            #s=250, marker='*',c='red',
            #edgecolor='black',
            #label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
from sklearn.datasets import make_blobs #自己創造數據
X,y = make_blobs(n_samples=150,  #隨機生成150個點 y是群
                  n_features=2,
                  centers=3,     #總共三群，密度高
                  cluster_std=0.5,#標準差
                  random_state=0)
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c='white',marker='o',edgecolor='black',s=50)
plt.grid()
plt.show()
```

```python
#計算群內誤差
from sklearn.cluster import KMeans
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0) #用kmeans的質心來算
    km. fit(X)
    distortions.append(km. inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel( 'Number of clusters')
plt.ylabel( 'Distortion')
plt.show() #會發現k=3發現轉折代表分三群最好
print("Distortion: %.2f " % km.inertia_) #群內誤差平⽅方和SSE
```

```python
#載入套件
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#讀取資料
data=load_wine()
X = pd.DataFrame(data['data'],columns=data['feature_names'])
Y = pd.DataFrame(data['target'],columns=['target'])
```

```python
#70%訓練30%測試
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state= 0)
```

```python
#標準化
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
X_train_std=sc_x.fit_transform(X_train)
X_test_std=sc_x.transform(X_test)
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
```

```python
from sklearn.linear_model import LogisticRegression
from matplotlib.colors import ListedColormap

lr=LogisticRegression(C=100.0,random_state=1)
lr.fit(X_train_pca,y_train)
```

```python
from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    markers=('s', 'x', 'o', '^', 'v')
    colors=('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
                                 
    x1_min, x1_max = X[:,0].min() - 1, X[:, 0].max()+ 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    z = z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
                    y=X[y == cl, 1],
                    alpha=0.8,
                    c=colors [idx],
                    marker=markers[idx],
                    label=cl,
                    edgecolor='black')
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(x_test [:, 0],
                    x_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100,
                    label='test set')
```

```python
plot_decision_regions(X_train_pca,y_train,classifier=lr)
plt.xlabel('sepal length [standardized]')
plt.ylabel('sepal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
#plt. savefig(' images. png', dpi=300)
plt.show()
```
