```python
from sklearn.datasets import load_diabetes

data=load_diabetes()
data.keys() #了解資料集裡面的鍵
data #了解資料集裡面的資料 => 發現data回傳array的資料是數值型 => Sex已經被數值化因此也能進行分析不用再轉換成數值
data['feature_names'] #了解欄位意義
```

```python
import pandas as pd
feature = pd.DataFrame(data['data'],columns=data['feature_names'])
target=pd.DataFrame(data['target'],columns=['target'])

print(feature.shape) #可以看出行數跟資料數 442列 10行  10種資料

df=pd.concat([feature,target],axis=1) #把feature跟target串聯在一起 互相對應
print(df.shape)#可以看出行數跟資料數 442列 11行
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6','target']

sns.pairplot(df[cols]) #看相關係數，用畫圖表示=>很難看懂 所以用熱度圖
plt.tight_layout()
plt.show()
```

```python
import numpy as np
cm=np.corrcoef(df[cols].values.T) #T是轉至 讓他變成5*5的 告訴他欄位才是變數，沒有執行的話會變成442*442(數值自己跑相關)

hm=sns.heatmap(cm,
              cbar=True,#圖片最右邊的條
              annot=True,#數值
              square=False,#圖片顯示為正方形
              fmt='.2f',#小數點取兩位
              annot_kws={'size':9},#調整字的大小
              yticklabels=cols,#x,y的標籤
              xticklabels=cols)  #熱度圖，能看變數間的相關係數  但是相關係數只能看出他的線性關係
plt.tight_layout()
plt.show()
```

由上方的熱度圖可以看出s1,s2兩個項目的相關高達0.9會導致模型共線性的問題，在經過預處理後應排除其中一個數，取一為代表

```python
from sklearn.model_selection import train_test_split
X,y = load_diabetes().data, load_diabetes().target
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state= 8)
```

```python
#未使用預處理
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm

slr=LinearRegression()
slr.fit(X_train,y_train)
print(slr.coef_)
y_train_pred=slr.predict(X_train)
y_test_pred=slr.predict(X_test)

print('MSE train:%.3f, test:%.3f' % (sm.mean_squared_error(y_train,y_train_pred),sm.mean_squared_error(y_test,y_test_pred)))
print('R^2 train:%.3f, test:%.3f' % (sm.r2_score(y_train,y_train_pred),sm.r2_score(y_test,y_test_pred)))
```

決定將s2排除，由於s1~s6的數值意義未知，因此經隨機選擇後選擇排除s2以s1作為代表。若以社會學研究為實際案例，研究者通常會根據兩個變數在研究假設中所佔的重要程度再決定在回歸模型中應該以哪個變數作為代表。

```python
#使用預處理
#根據熱度圖發現s2與s1高度相關，排除s2
data=load_diabetes()
X = pd.DataFrame(data['data'],columns=data['feature_names'])
X.drop(columns=['s2'],inplace=True) #排除s2
y = load_diabetes().target
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state= 8)
```

```python
#使用預處理
slr=LinearRegression()
slr.fit(X_train,y_train)
print(slr.coef_)
y_train_pred=slr.predict(X_train)
y_test_pred=slr.predict(X_test)

print('MSE train:%.3f, test:%.3f' % (sm.mean_squared_error(y_train,y_train_pred),sm.mean_squared_error(y_test,y_test_pred)))
print('R^2 train:%.3f, test:%.3f' % (sm.r2_score(y_train,y_train_pred),sm.r2_score(y_test,y_test_pred)))
```

比較兩者的係數，首先在解釋力 R下降了0.003 然而從迴歸係數中可以看出很大的差異，例如s1的強度原本是高達-1043但在排除s2後變成-292，其他數值也
有所變動。經過資料預處理後能反映出正確的迴歸係數，不會因為變數互相影響導致某些變數的強度有膨脹導致模型偏誤的問題

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import Ridge
```

```python
#ridge是為了要達到模型盡量簡單，但保持好的解釋力
ridge=Ridge(alpha=10).fit(X_train,y_train) #模型前面*一個alpha值來加權強度  alpha帶 1,10,0.1測試  alpha值越小越接近線性回歸的結果
print(ridge.coef_)

y_train_pred=ridge.predict(X_train)
print('R^2 train:%.3f' % (sm.r2_score(y_train,y_train_pred))) #(1,0.433)  如果在意的是模型的預測能力的準確度就要用ridge
#可以提高alpha值來解決overfitting，同時也能保有模型複雜度 但也會降低訓練集的分數
```

```python
ridge_t=Ridge(alpha=10).fit(X_test,y_test)
print(ridge.coef_)

y_test_pred=ridge.predict(X_test)
print('R^2  test:%.3f' % (sm.r2_score(y_test,y_test_pred))) #(1,0.433)
```

```python
from sklearn.linear_model import Lasso #公式變成用絕對值，所以最多會收斂到0
Lasso=Lasso(alpha=0.1).fit(X_train,y_train) #alpha=1,0.1,0.001
print(ridge.coef_)

y_train_pred=Lasso.predict(X_train)
print('R^2 train:%.3f' % (sm.r2_score(y_train,y_train_pred))) #0.362,0.519,0.530
```

```python
Lasso_t=Ridge(alpha=0.1).fit(X_test,y_test)#alpha=1,0.1,0.001 只用1應該會有其他係數為0 用0.1會取7個特徵
print(ridge.coef_)

y_test_pred=Lasso.predict(X_test)
print('R^2  test:%.3f' % (sm.r2_score(y_test,y_test_pred))) #0.366,0.460,0.480
```

Lasso會犧牲模型正確性，但特徵太多，只有一小部分重要就可用Lasso，用解釋模型Lasso比較好理解因為採取特徵少

```python
#Elastic Net彈性網=>把ridge跟lasso的係數都放進去
from sklearn.linear_model import ElasticNet
elanet=ElasticNet(alpha=1, l1_ratio=0.5).fit(X_train,y_train) #alpha=a+b   L1_ratio=a/a+b
print(elanet.coef_)

y_train_pred=elanet.predict(X_train)
print('R^2 train:%.3f' % (sm.r2_score(y_train,y_train_pred)))
```

這個算法的問題在於，要處理異常值和共線性的問題(預測因子權重被扭曲)。因為每個點的比重一樣所以有異常值會使模型受影響
曲線趨勢問題=>要先觀察目標變數與變數的關係，可能不是直線關係(可以取log解決)，並且預測因子和結果並不一定是因果關係

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

iris=load_iris()

feature=pd.DataFrame(iris['data'],columns=iris['feature_names'])
target=pd.DataFrame(iris['target'],columns=['class'])

data=pd.concat([feature,target],axis=1)
df=data[data['class'] != 2] #只考慮0和1的資料
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
g=sns.FacetGrid(df,hue="class",size=5)
g.map(plt.scatter,"sepal length (cm)","sepal width (cm)")
g.add_legend()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X=df.iloc[:,:2].values
y=df.iloc[:,4].values

sc=StandardScaler()
sc.fit(X)
X_std=sc.transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
from matplotlib.colors import ListedColormap

lr=LogisticRegression(C=100.0,random_state=1)
lr.fit(X_std,y)

print(lr.coef_)
print(lr.intercept_)
```

```python
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    markers=('s', 'x', 'o', '^', 'v')
    colors=('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
                                 
    x1_min,x1_max=X[:,0].min()-1,X[:, 0].max()+1
    x2_min,x2_max=X[:,1].min()-1,X[:,1].max()+1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    z = z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl,0],
                    y=X[y==cl,1],
                    alpha=0.8,
                    c=colors [idx],
                    marker=markers[idx],
                    label=cl,
                    edgecolor='black')
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(x_test [:, 0],
                    x_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100,
                    label='test set')
```

```python
plot_decision_regions(X_std, y, classifier=lr)
plt.xlabel('sepal length [standardized] ')
plt.ylabel('sepal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
#plt. savefig(' images. png', dpi=300)
plt.show()
```
