```python
from sklearn.datasets import make_blobs #自己創造數據
X,y = make_blobs(n_samples=150,  #隨機生成150個點
                  n_features=2,
                  centers=3,     #總共三群，密度高
                  cluster_std=0.5,#標準差
                  random_state=0)
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c='white',marker='o',edgecolor='black',s=50)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,
         init="random",
         n_init=10,
         max_iter=300,
         tol=1e-04,
         random_state=0)
y_km = km.fit_predict(X) #訓練出來的結果
```

```python
plt.scatter(X[y_km == 0,0],
            X[y_km == 0,1],
            s=50,c='lightgreen',
            marker='s',edgecolor='black',
            label='cluster 1')
plt.scatter(X[y_km == 1,0],
            X[y_km == 1,1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='cluster 2')
plt.scatter(X[y_km == 2,0],
            X[y_km == 2,1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*',c='red',
            edgecolor='black',
            label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,
         init="k-means++",
         n_init=10,
         max_iter=300,
         tol=1e-04,
         random_state=0)
y_km = km.fit_predict(X) 
```

```python
plt.scatter(X[y_km == 0,0],
            X[y_km == 0,1],
            s=50,c='lightgreen',
            marker='s',edgecolor='black',
            label='cluster 1')
plt.scatter(X[y_km == 1,0],
            X[y_km == 1,1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='cluster 2')
plt.scatter(X[y_km == 2,0],
            X[y_km == 2,1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*',c='red',
            edgecolor='black',
            label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
Complete = linkage(X, method = 'complete', metric = 'euclidean')
dendrogram (Complete)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel('Cluster distance')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
Single = linkage(X, method = 'single', metric = 'euclidean')
dendrogram (Single)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel('Cluster distance')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
Average = linkage(X, method = 'average', metric = 'euclidean')
dendrogram (Average)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel('Cluster distance')
plt.show()
```

```python
from scipy.cluster.hierarchy import ward
from scipy.cluster.hierarchy import dendrogram
linkage=ward(X)
dendrogram(linkage)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel('Cluster distance')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
X,y=make_moons(n_samples=200,noise=0.05,random_state=0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.2,min_samples=5,metric="euclidean")
y_db=db.fit_predict(X)
```
